# 4.1.1 Numpyを用いた行列演算



```python
# numpy ライブラリをインポート
import numpy as np

# --- 基礎的な配列処理 ---
# リストから配列を作る
my_list = [1, 2, 3, 4, 5]
my_array = np.array(my_list)
print("my_list から作られた Numpy 配列:\n", my_array)
print("my_array の型:", my_array.dtype)

# ある特定の型を持つ Numpy 配列を作る
my_array_float = np.array(my_list, dtype=float)
print("float 型の Numpy 配列:\n", my_array_float)
print("my_array の型:", my_array_float.dtype)

# すべての要素が0の配列を作る
zeros_array = np.zeros(5) # 要素数 5 の配列
print("要素数5のゼロベクトル:\n", zeros_array)

# すべての要素が1の配列を作る
ones_array = np.ones((2, 3))  # 2x3 の配列
print("2x3の要素数を持つ all-one 行列:\n", ones_array)

# 等差数列を作る
sequence_array = np.arange(0, 10, 2)  # 0から開始し（終点を含まずに）10までを2刻みで含む配列
print("0～10までの、公差が2の等差数列（終点は含まない）:\n", sequence_array)

# 指定された区間を等分した配列を作る
linspace_array = np.linspace(0, 1, 5)  # 0から1までを、始点と終点も含めて5等分した結果を含む配列
print("0～1までの、交差が (1 - 0)/5 の等差数列（終点も含む）:\n", linspace_array)

# 単位行列を作る
identity_matrix = np.eye(3)  # 3x3 の単位行列
print("3x3の単位行列:\n", identity_matrix)

# 乱数を含む配列を作る
random_array = np.random.rand(3, 3) # 3x3 の、一様分布 U(0, 1) からサンプリングされた要素を持つ配列
print("乱数配列:\n", random_array)

# --- 配列の情報を得る ---
print("形状:", my_array.shape)
print("データ型:", my_array.dtype)
print("次元の数:", my_array.ndim)
print("要素数:", my_array.size)

# --- 配列の参照とスライシング ---
print("index が2の要素:\n", my_array[2])
print("index が1～3の要素（ただし index = 3 は除外）:\n", my_array[1:3])
print("index が偶数の要素:\n", my_array[::2])

# --- 行列演算 ---
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("要素ごとの和:\n", arr1 + arr2)
print("要素ごとの差:\n", arr1 - arr2)
print("要素ごとの積:\n", arr1 * arr2)
print("要素ごとの商:\n", arr1 / arr2)
print("行列積:\n", np.dot(arr1, arr2)) # or arr1 @ arr2

# 基本的な関数
print("要素ごとの平方根:\n", np.sqrt(arr1))
print("要素ごとの exp:\n", np.exp(arr1))
print("総和:", np.sum(arr1))
print("平均:", np.mean(arr1))
print("最大値:", np.max(arr1))
print("最小値:", np.min(arr1))

# 配列の形状変更 (reshape)
reshaped_array = my_array.reshape(1, 5) # my_array を 1x5 の行列に変更
print("形状変更後の配列:\n", reshaped_array)
print("形状:", reshaped_array.shape)
print("次元の数:", reshaped_array.ndim)
```

# 4.1.2 PyTorchを用いた行列演算

```python
# torch ライブラリをインポート
import torch

# --- 基礎的な配列処理 ---
# リストから tensor を作る
my_list = [1, 2, 3, 4, 5]
my_tensor = torch.tensor(my_list)  # デフォルトは float32 で生成される
print("my_list から作られた PyTorch 配列 (tensor):\n", my_tensor)
print("my_tensor の型:", my_tensor.dtype)

# ある特定の型を持つ tensor を作る
my_tensor_int = torch.tensor(my_list, dtype=torch.int64)
print("int 型の tensor:\n", my_tensor_int)
print("my_tensor の型:", my_tensor_int.dtype)

# すべての要素が0の tensor を作る
zeros_tensor = torch.zeros(5, dtype=torch.float32) # 要素数 5 の tensor
print("要素数5のゼロベクトル:\n", zeros_tensor)

# すべての要素が1の tensor を作る
ones_tensor = torch.ones((2, 3), dtype=torch.float32) # 2x3 の tensor
print("2x3 の all-one tensor:\n", ones_tensor)

# 等差数列を作る
sequence_tensor = torch.arange(0, 10, 2, dtype=torch.float32) # 0から開始し（終点を含まずに）10までを2刻みで含む tensor
print("0～10までの、公差が2の等差数列（終点は含まない）:\n", sequence_tensor)

# 指定された区間を等分した配列を作る
linspace_tensor = torch.linspace(0, 1, 5, dtype=torch.float32) # 0から1までを、始点と終点も含めて5等分した結果を含む tensor
print("0～1までの、交差が (1 - 0)/5 の等差数列（終点も含む）:\n", linspace_tensor)

# 単位行列を作る
identity_matrix = torch.eye(3, dtype=torch.float32)  # 3x3 の単位行列
print("3x3の単位行列:\n", identity_matrix)

# 乱数を含む配列を作る
random_tensor = torch.rand(3, 3) # # 3x3 の、一様分布 U(0, 1) からサンプリングされた要素を持つ配列
print("乱数 tensor:\n", random_tensor)

# --- Tensor の情報を得る ---
print("形状:", my_tensor.shape)
print("データの型:", my_tensor.dtype)
print("次元の数:", my_tensor.ndim)
print("要素数:", my_tensor.numel())

# --- Tensor の参照とスライシング ---
print("index が2の要素:\n", my_tensor[2])
print("index が1～3の要素（ただし index = 3 は除外）:\n", my_tensor[1:3])
print("index が偶数の要素:\n", my_tensor[::2])

# --- 行列演算 ---
tensor1 = torch.tensor([1, 2, 3], dtype=torch.float32)
tensor2 = torch.tensor([4, 5, 6], dtype=torch.float32)

print("要素ごとの和:\n", tensor1 + tensor2)
print("要素ごとの差:\n", tensor1 - tensor2)
print("要素ごとの積:\n", tensor1 * tensor2)
print("要素ごとの商:\n", tensor1 / tensor2)
print("行列積:\n", torch.dot(tensor1, tensor2))

# 基本的な関数
print("要素ごとの平方根:\n", torch.sqrt(tensor1))
print("要素ごとの exp:\n", torch.exp(tensor1))
print("総和:", torch.sum(tensor1))
print("平均:", torch.mean(tensor1))
print("最大値:", torch.max(tensor1))
print("最小値:", torch.min(tensor1))

# Tensor の形状変更 (reshape)
reshaped_tensor = my_tensor.reshape(1, 5)  # 1x5 tensor に変更
print("形状変更後の tensor:\n", reshaped_tensor)
print("形状:", reshaped_tensor.shape)
print("次元の数:", reshaped_tensor.ndim)
```

# 4.1.3 matplotlibを用いたデータ可視化

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 正弦波を 0 ～ 4 pi の範囲でプロットしてみる ---
x = np.linspace(0, 4*np.pi, 100)  # 0から10までを100等分した数列を生成
y = np.sin(x)  # xのサイン波
plt.plot(x, y) # (x, y) の曲線をプロット
plt.xlabel("x") # x軸のラベルを設定
plt.ylabel("y") # y軸のラベルを設定
plt.title("sin wave") # グラフのタイトルを設定
plt.grid(True) # グリッド線の表示
plt.show() # プロットの表示

#複数のグラフを一つの図に描画
y2 = np.cos(x)
plt.plot(x, y, label="sin(x)") # labelオプションで凡例を指定できる
plt.plot(x, y2, label="cos(x)", color="red") # colorオプションで色を指定できる
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin wave and cos wave")
plt.legend() # 凡例を表示
plt.grid(True)
plt.show()

# --- 100個の一様乱数をプロットしてみる ---
x = np.random.rand(100)
y = np.random.rand(100)
plt.scatter(x, y) # (x, y) の散布図
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatter plots")
plt.show()

# --- 1000個の正規乱数をプロットしてみる ---
data = np.random.randn(1000) # 正規分布から1000個の乱数を生成
plt.hist(data, bins=30)  # 30個のビンでヒストグラムを作成
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()

# --- 4カテゴリの棒グラフをプロットしてみる ---
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values) # categories に関する棒グラフ
plt.xlabel("Label")
plt.ylabel("Value")
plt.title("Bar plot")
plt.show()
```
